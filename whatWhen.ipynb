```python
from matplotlib import pyplot as plt
from matplotlib.table import Table

# Define the cheat sheet data
models = [
    "Linear Regression",
    "Ridge / Lasso Regression",
    "Logistic Regression",
    "K-Nearest Neighbors (KNN)",
    "Decision Tree",
    "Random Forest",
    "Gradient Boosting (e.g., XGBoost)",
    "Support Vector Machine (SVM)",
    "Naive Bayes",
    "K-Means Clustering",
    "PCA (Dimensionality Reduction)",
    "Neural Networks"
]

problem_types = [
    "Regression",
    "Regression",
    "Binary Classification",
    "Classification / Regression",
    "Classification / Regression",
    "Classification / Regression",
    "Classification / Regression",
    "Classification",
    "Classification",
    "Clustering (Unsupervised)",
    "Dimensionality Reduction",
    "Classification / Regression"
]

best_for = [
    "Continuous values with linear patterns",
    "High-dimensional or noisy data",
    "2-class problems like spam detection",
    "Small, clean datasets",
    "Interpretable models, non-linear data",
    "Mixed-type tabular data",
    "High-performance competitive modeling",
    "Small datasets, clear class margins",
    "Text data like spam, sentiment",
    "Grouping unlabeled data",
    "Removing noise, speeding up training",
    "Complex non-linear patterns"
]

why_use = [
    "Simple, fast, and interpretable",
    "Adds regularization to prevent overfitting",
    "Good baseline model, fast and explainable",
    "Intuitive and non-parametric",
    "Easy to visualize and handle non-linear data",
    "Reduces overfitting and works well out of the box",
    "Handles complex patterns, top-tier performance",
    "Effective in high-dimensions with margin maximization",
    "Fast and works well for text classification",
    "Quick, intuitive grouping of similar data",
    "Speeds up training, useful for visualization",
    "Very powerful, but needs lots of data and tuning"
]

# Plotting the cheat card
fig, ax = plt.subplots(figsize=(12, 9))
ax.set_axis_off()
table = Table(ax, bbox=[0, 0, 1, 1])

n_rows = len(models) + 1
n_cols = 4
col_widths = [0.2, 0.2, 0.3, 0.3]

# Header
headers = ["Model", "Problem Type", "Best For", "Why Use It?"]
for i, header in enumerate(headers):
    table.add_cell(0, i, width=col_widths[i], height=0.05, text=header, loc='center', facecolor='#cccccc')

# Fill rows
for row in range(len(models)):
    table.add_cell(row + 1, 0, width=col_widths[0], height=0.05, text=models[row], loc='left')
    table.add_cell(row + 1, 1, width=col_widths[1], height=0.05, text=problem_types[row], loc='left')
    table.add_cell(row + 1, 2, width=col_widths[2], height=0.05, text=best_for[row], loc='left')
    table.add_cell(row + 1, 3, width=col_widths[3], height=0.05, text=why_use[row], loc='left')

ax.add_table(table)
plt.tight_layout()
plt.show()
```
